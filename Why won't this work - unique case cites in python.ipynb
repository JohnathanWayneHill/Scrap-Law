```python

#get unique case links
#All links
opinion = requests.get("https://scholar.google.com/scholar_case?case=9151214196953452755&q=ARTIFICIAL&hl=en&scisbd=2&as_sdt=4,176")
soup = bs(opinion.content, 'html.parser')

all_links = soup.select('a')
old_links = soup.select("a.gsl_co_link")
old_link_text = [i.text for i in old_links]
old_link_links = [link['href'] for link in old_links]
#This gets every case I don't hve a link to. The link didn't work for a 2007 federal district court case that I was able to find
#elsewhere.

#################################################
######## Filtering Old Links
links = [link['href'] for link in all_links] # extracting links 
filtered_link_list = []
# This removes all of the old case links
for i in links:
    if i not in old_link_links: 
        filtered_link_list.append(i)
        print(i)

# Fltering out non case links
def just_case_links(a): # what's being done with 
    ''' 
    Describe function 
    '''
    pattern = re.compile(r'scholar_case\?case') # ...
    matches = pattern.findall(a) 
    return matches[0] 
    ''' was here but commmented: 
    for match in matches:
            return match
    '''

case_links = []
for i in filtered_link_list:
    if type(just_case_links(i)) == str:
        case_links.append(i)

unique_case_links = []
for i in case_links:
    if i not in unique_case_links: 
        unique_case_links.append(i)

print(len(case_links))
print(len(unique_case_links))
```
